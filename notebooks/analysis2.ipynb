### Graph 1 Parks around the world

```python
parks1_clean = (parks1
                .dropna(axis=0)
                .sort_values(by = "city", ascending = True)
               )
parks1_clean.head()
```

```python
parks2_clean = (parks1_clean
                .drop(["park.key"], axis = 1)
                .dropna(axis = 0)
               )
```

```python
parks2_clean.columns
```

```python
parks3_clean = parks2_clean.rename({'park.name': "Park Name", 'park.alias': "Park Alias", 'city': "City", 'state': "State", 'country': "Country"}, axis = 1).reset_index().drop(['index'], axis = 1)
```

```python
parks3_clean.head()
```

```python
parks4_clean = (parks3_clean
                   .drop(["Country"], axis = 1)
                  )
```

```python
parks4_clean.describe()
```

```python
parks4_clean["State"].value_counts()
```

```python
labels = ['OH', 'CA', 'IL', 'NY', 'PA', 'MA', 'MO', 'MI', 'IN', 'TX', 'QC', 'MD', 'IA', 'FL', 'CT', 'AZ', 'WA', 'ON', 'DC']
```

```python
final_df = parks4_clean.loc[parks4_clean['State'].isin(['OH', 'CA', 'IL', 'NY', 'PA', 'MA', 'MO', 'MI', 'IN', 'TX', 'QC', 'MD', 'IA', 'FL', 'CT', 'AZ', 'WA', 'ON', 'DC']
)]
```

###### find a way to make a histogram graph of how many stadiums are in each state. Shows where watching baseball is more popular.

```python
plt.style.use('tableau-colorblind10')
fig = plt.figure(figsize=(11,11))
plt.pie(list(final_df['State'].value_counts()), autopct='%1.1f%%',labels = labels, wedgeprops={'edgecolor':'black'})
plt.title('Parks located around the World used in the MLB')
plt.tight_layout()
plt.show()
```

### Graph 2 Players with the most games played Fielding from 2000 >

```python
fielding1_clean = (fielding1
                .sort_values(by = "G", ascending = False)
               )
fielding1_clean
```

```python
recentfielding_clean = fielding1_clean[fielding1_clean["yearID"] > 1999]
recentfielding_clean.head()
```

```python
fielding2_clean = (recentfielding_clean
                   .drop(["stint", "WP", "GS", "ZR", "lgID", "E", "DP", "InnOuts", "PO", "A", "PB", "SB", "CS", "yearID", "POS"], axis = 1)
                  )
```

```python
fielding3_clean = fielding2_clean.rename({'playerID': "Player ID", 'yearID': "Year", 'teamID': "Team", 'G': "Games Played"}, axis = 1).reset_index().drop(['index'], axis = 1)
```

```python
fielding3_clean.head()
```

```python
fielding4_clean = (fielding3_clean.dropna(axis = 0))
```

```python
fielding5_clean = fielding4_clean[fielding4_clean["Games Played"] > 150]
fielding5_clean
```

```python
fielding5_clean.describe()
```

```python
fielding5_clean['Team'].value_counts()
```

```python
labels = ['PHI', 'SEA', 'ATL', 'BAL', 'LAN', 'NYA', 'CHN', 'KCA', 'OAK', 'FLO', 'COL', 'TEX', 'MIL', 'CIN', 'ARI', 'SDN', 'HOU', 'TOR', 'CHA', 'DET', 'BOS', 'WAS', 'PIT', 'SLN', 'NYN', 'CLE', 'MON', 'SFN', 'ANA', 'LAA', 'MIA', 'TBA', 'MIN']
```

```python
final2_df = fielding5_clean.loc[fielding5_clean['Team'].isin(['PHI', 'SEA', 'ATL', 'BAL', 'LAN', 'NYA', 'CHN', 'KCA', 'OAK', 'FLO', 'COL', 'TEX', 'MIL', 'CIN', 'ARI', 'SDN', 'HOU', 'TOR', 'CHA', 'DET', 'BOS', 'WAS', 'PIT', 'SLN', 'NYN', 'CLE', 'MON', 'SFN', 'ANA', 'LAA', 'MIA', 'TBA', 'MIN']
)]
```

```python
plt.style.use('tableau-colorblind10')
fig = plt.figure(figsize=(11,11))
plt.pie(list(final2_df['Team'].value_counts()), autopct='%1.1f%%',labels = labels, wedgeprops={'edgecolor':'black'})
plt.title('What Teams in the MLB have the majority of Veteran players (>150 games played)')
plt.tight_layout()
plt.show()
#This graph shows which teams have the highest amounts of players that have played more than 150 games.
```
