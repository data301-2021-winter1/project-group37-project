### Graph 2 Players with the most games played Fielding from 2000 >

```python
fielding1_clean = (fielding1
                .sort_values(by = "G", ascending = False)
               )
fielding1_clean
```

```python
recentfielding_clean = fielding1_clean[fielding1_clean["yearID"] > 1999]
recentfielding_clean.head()
```

```python
fielding2_clean = (recentfielding_clean
                   .drop(["stint", "WP", "GS", "ZR", "lgID", "E", "DP", "InnOuts", "PO", "A", "PB", "SB", "CS", "teamID", "yearID", "POS"], axis = 1)
                  )
#will drop putouts, assists, InnOuts, CS and SB later by merging factors and position as all will be catchers
```

```python
fielding3_clean = fielding2_clean.rename({'playerID': "Player ID", 'yearID': "Year", 'teamID': "Team", 'G': "Games Played"}, axis = 1).reset_index().drop(['index'], axis = 1)
```

```python
fielding3_clean.head()
```

```python
fielding4_clean = (fielding3_clean.dropna(axis = 0))
fielding4_clean
```

```python
fielding5_clean = fielding4_clean[fielding4_clean["Games Played"] > 161]
fielding5_clean
```
