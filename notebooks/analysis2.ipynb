```python
parks1 = pd.read_csv("../data/raw/Parks.csv")
parks1.head()
```

```python
parks1_clean = (parks1
                .dropna(axis=0)
                .sort_values(by = "park.key", ascending = True)
               )
parks1_clean.head()
```

```python
parks2_clean = (parks1_clean
                .drop(["park.key"], axis = 1)
                .dropna(axis = 0)
               )
parks2_clean.head()
```

```python
parks2_clean.columns
```
